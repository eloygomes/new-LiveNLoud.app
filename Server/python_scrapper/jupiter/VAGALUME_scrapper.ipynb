```python
import re
import requests
from bs4 import BeautifulSoup

MOBILE_UA = (
    "Mozilla/5.0 (iPhone; CPU iPhone OS 14_0 like Mac OS X) "
    "AppleWebKit/605.1.15 (KHTML, like Gecko) Version/14.0 "
    "Mobile/15E148 Safari/604.1"
)
# URL = "https://www.vagalume.com.br/dire-straits/sultans-of-swing.html"
URL = "https://www.vagalume.com.br/the-beatles/help.html"

def clean(text: str) -> str:
    text = text.replace("\r", "")
    text = re.sub(r"\n{3,}", "\n\n", text)
    return text.strip()

def scrape_vagalume(url: str):
    headers = {
        "User-Agent": MOBILE_UA,
        "Accept-Language": "pt-BR,pt;q=0.9,en-US;q=0.8,en;q=0.7",
    }
    resp = requests.get(url, headers=headers, timeout=30)
    resp.raise_for_status()

    soup = BeautifulSoup(resp.text, "html.parser")

    title = soup.select_one("#lyricContent h1")
    artist = soup.select_one("#lyricContent h2")
    lyrics_div = soup.select_one("#lyrics")

    lyrics = None
    if lyrics_div:
        # converte <br> em \n e tira anúncios/iframes internos
        for ad in lyrics_div.find_all(["div", "iframe"], recursive=False):
            ad.decompose()
        lyrics = clean(lyrics_div.get_text("\n", strip=True))

    return {
        "title": title.get_text(strip=True) if title else None,
        "artist": artist.get_text(strip=True) if artist else None,
        "lyrics": lyrics,
    }

if __name__ == "__main__":
    data = scrape_vagalume(URL)
    print("Música:", data["title"])
    print("Artista:", data["artist"])
    print("\nLetra:\n")
    print(data["lyrics"])
```
