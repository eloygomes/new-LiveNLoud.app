```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup

# configura o Chrome para emular mobile (via user-agent)
chrome_options = Options()
chrome_options.add_argument(
    "user-agent=Mozilla/5.0 (iPhone; CPU iPhone OS 14_0 like Mac OS X) "
    "AppleWebKit/605.1.15 (KHTML, like Gecko) Version/14.0 Mobile/15E148 Safari/604.1"
)
# opcional: abrir sem janela
chrome_options.add_argument("--headless")

driver = webdriver.Chrome(options=chrome_options)
try:
    # CHORDS
    driver.get("https://tabs.ultimate-guitar.com/tab/the-beatles/help-chords-17269")
    # driver.get("https://tabs.ultimate-guitar.com/tab/1448029")
    # driver.get("https://tabs.ultimate-guitar.com/tab/bob-dylan/hurricane-chords-14871")

    # TABS
    # driver.get("https://tabs.ultimate-guitar.com/tab/the-beatles/help-tabs-93482")
    
    # BASS
    # driver.get("https://tabs.ultimate-guitar.com/tab/the-beatles/help-bass-17471")

    #DRUMS
    # driver.get("https://tabs.ultimate-guitar.com/tab/the-beatles/help-drums-865752")
    
    
    
    # aguarda o carregamento completo se necessário (ex: WebDriverWait)

    soup = BeautifulSoup(driver.page_source, "html.parser")
    main = soup.find("body")
    # print(main)

    # exemplo: pegar o título usando a classe observada no mobile
    title = main.find("h1", class_="tabHeader-h1").get_text(strip=True)
    

    info_list = soup.find("ul", class_="tabHeader-info")

    # TUNNING
    if info_list:
        for item in info_list.find_all("li", class_="tabHeader-item"):
            label = item.find("span", class_="tabHeader-name")
            if label and label.get_text(strip=True) == "Tuning:":
                tuning = item.get_text(strip=True).replace("Tuning:", "").strip()
                break

    
    # KEY
    if info_list:
         for item in info_list.find_all("li", class_="tabHeader-item"):
            label = item.find("span", class_="tabHeader-name")
            if label and label.get_text(strip=True) == "Key:":
                key = item.get_text(strip=True).replace("Key:", "").strip()
                break


    cifra = main.find("pre", class_="extra")


    print(f"title",title)
    print(f"tuning",tuning)
    print(f"key",key)
    print(f"cifra:", cifra)



finally:
    driver.quit()
```
